```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_feather("/work/sdaigler/smhtt_ul/training_datasets/folds/fold0_training.feather")
is_data = (df.Labels.is_data == 1)
is_not_data = (df.Labels.is_data == 0)
is_nominal = (df.Nominal.cut == 1)

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
bins = np.linspace(0, 600, 101)
df.Nominal.variables.pt_1[is_data & is_nominal].hist(bins=bins, ax=ax, label="is_data", histtype="step")
df.Nominal.variables.pt_1[is_not_data & is_nominal].hist(bins=bins, ax=ax, label="is_not_data", histtype="step", weights=df.Nominal.weight[is_not_data & is_nominal])
ax.legend()
```

```python
count_is_data = is_data.sum()
count_is_not_data = is_not_data.sum()

print(f"Number of is_data events: {count_is_data}")
print(f"Number of is_not_data events: {count_is_not_data}")

count_is_data_and_is_nominal = (is_data & is_nominal).sum()
count_is_not_data_and_is_nominal = ((is_not_data & is_nominal) * df.Nominal.weight[is_not_data & is_nominal]).sum()

print(f"Number of events with is_data and is_nominal: {count_is_data_and_is_nominal}")
print(f"Number of events with is_not_data and is_nominal: {count_is_not_data_and_is_nominal}")
```
